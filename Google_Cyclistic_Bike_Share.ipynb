## **Data Merging**

> On this project, **multiple datasets will be merged into 1 dataset because 1 CSV dataset file only records bike trips for one month**. Whereas for analysis, it requires a **historical dataset of bike trips for the last 12 months or from October 2021 to September 2022**. Therefore, dataset merging is necessary.

> **Multiple datasets with CSV file format will be merged with the `Union All` function using PostgresSQL** where previously the datasets have been stored in a database. **The datasets to be merged have the same number of fields (columns) and field names (columns).**

> Note: The `Union All` function in SQL is used to join two or more tables into one table vertically, by placing records (rows) from the first table followed by records (rows) from the second table and so on. The `Union All` function only joins data from tables that have the same number of fields (columns) and does not remove duplicate data.

> Below is the `Union All` query statement to join 12 tables:

```

(
```

````python
df_trips[['started_at', 'ended_at']] = df_trips[['started_at', 'ended_at']].astype('datetime64[ns]')
df_trips = df_trips.rename({'member_casual': 'user_type'}, axis='columns')
df_trips = df_trips.astype({'rideable_type':'category', 'user_type':'category'})
df_trips.info()
````

> The data has been cleaned and then proceeds to the data manipulation stage.

## **Data Manipulation**

> Data manipulation is the process of changing or altering data to make it more **understandable and organized**. The data manipulation that will be conducted is **extracting or adding data and filtering information based on the analysis requirements**.

> From this dataset with 13 columns/variables, we need new columns that show information about the **length of each trip, which days of the week the rider frequently trips, and what time the rider starts the trip**. These metrics will be an **important insight** to answer the question **"How do annual members and casual riders use Cyclistic bikes differently?"**.

> **Creating a new variable**, **`ride_duration`**, which shows the **duration** of the user's trip in **minutes** using the `.insert()` function, by calculating the **time difference between the `ended_at` and `started_at` variables**. Then sort the values of the `ride_duration` variable in **ascending order** to see what the **lowest value** is or identify any errors such as **negative values**.

````python
df_trips.insert(loc=4, column="ride_duration", value= df_trips['ended_at'] - df_trips['started_at'])
df_trips['ride_duration'] = df_trips['ride_duration'] / pd.Timedelta(minutes=1)
df_trips.sort_values(by = 'ride_duration')
````

> It can be seen that many rows contain **negative values**. The error occurs because the "end time" is earlier than the "start time" in each row. In addition, there is an **extreme value** in the `ride_duration` variable, which is **40705 minutes** or **22 days** with the user type being **casual riders**. This is unusual considering that **casual riders are customers who purchase single-ride or full-day tickets**, which means that the maximum **use of the bike is one day** and the program of the Cyclistic company is **short-term rentals throughout the city of Chicago**. Perhaps these values are **outliers**, so we need to check for **outliers** using **boxplots**.

````python
sns.boxplot(data = df_trips, x ='user_type', y = 'ride_duration', order = ['member', 'casual'])
plt.show()
````

> **Outliers** in the `ride_duration` variable for **casual riders** are **numerous**. Since there is **no information** on the **limit** of **maximum duration** of **bike usage** in **Cyclistic program**, the **median** value will be used instead of the **mean** value to determine the difference between **casual riders** and **annual members** based on **trip duration**. That's because the **mean** value is strongly **influenced** by **outliers** while the **median** value is **not**, so the **median** value will be more **accurate** for our **analysis**.

> Of course, the `ride_duration` variable will be **filtered**. **Values below 1 minute including negative values will not be used** because it is **impossible** for the **ride duration** to be **0 minutes to negative**.

````python
df_trips = df_trips[df_trips['ride_duration'] >= 1]
df_trips = df_trips.reset_index(drop=True)

df_trips.head(4)
````

> The new columns `ride_hour`, `ride_day`, `ride_month`, `ride_year`, and `ride_season` are added by extracting the value of `started_at`. These columns are created to make the data **more understandable** and as **a metric to find out the times when riders often start their rides**.

> **Season information** in the City of **Chicago** for **October 2021 to September 2022** was obtained from [**Weather Spark**](https://id.weatherspark.com/h/y/14091/2021/Cuaca-Historis-selama-2021-di-Chicago-Illinois-Amerika-Serikat) and here are the details:
* **Summer**: June, July, and August.
* **Winter**: December, January, and February.
* **Spring**: March, April, and May.
* **Autumn**: September, October, and November.

````python
df_trips.insert(loc=5, column="ride_hour", value= df_trips['started_at'].dt.hour)

day = [ "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
df_trips.insert(loc=6, column="ride_day", value= df_trips['started_at'].dt.day_name())
df_trips['ride_day'] = df_trips['ride_day'].astype(CategoricalDtype(categories=day, ordered=False))

month = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df_trips.insert(loc=7, column="ride_month", value= df_trips['started_at'].dt.month_name())
df_trips['ride_month'] = df_trips['ride_month'].astype(CategoricalDtype(categories=month, ordered=False))

df_trips.insert(loc=8, column="ride_year", value= df_trips['started_at'].dt.year)

seasons_dict = {'Winter': ['December', 'January', 'February'],
                'Spring': ['March', 'April', 'May'],
                'Summer': ['June', 'July', 'August'],
                'Autumn': ['September', 'October', 'November']}
df_trips.insert(loc=9, column="ride_season", value= df_trips['ride_month'].map(
    {month: season for season, months in seasons_dict.items() for month in months}))

df_trips.head(4)
````

## **Total Trips**

````python
def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: lime' if v and i != ('All', '') else '' for i,v in zip(s.index, is_max)]
````

````python
df_pv1 = pd.pivot_table(df_trips,
                       index='user_type',
                       values='ride_id',
                       aggfunc=['count'],
                       margins=True,
                       margins_name='All',
                       observed=True)

total_count = df_pv1.loc['All', ('count', 'ride_id')]
df_pv1['% of count'] = df_pv1[('count', 'ride_id')].apply(lambda x: "{:.2%}".format(x / total_count))
df_pv1.style.background_gradient(cmap='YlGn', axis=0, subset=pd.IndexSlice[df_pv1.index[:-1], df_pv1.columns[:1]])
````

It shows that **annual members** represent **the majority of bike trips** using **Cyclistic Bike-Share** during **October 2021** to **September 2022**.

## **Total Monthly & Seasonal Trips**

> Having a total of **5.71 million trips** during **October 2021 to September 2022**, it is important to know the **busiest month and season** of Cyclistic trips to assist in determining **marketing strategies based** on **Cyclistic user behavior**.

### **Seasonality**

````python
df_pv3 = pd.pivot_table(df_trips,
              index = ['ride_season'],
              columns = ['user_type'],
              values = 'ride_id',
              aggfunc = ['count'],
              margins = True)
df_pv3.style.background_gradient(cmap='YlGn', axis=0, subset=pd.IndexSlice[df_pv3.index[:-1], df_pv3.columns[:3]])
````

> **Insight**
* From **October 2021 to September 2022**, **Cyclistic's total trips** were **highest** in the **summer** as well as for **each user (annual members and casual riders)**. According to [**Weather Spark**](https://id.weatherspark.com/s/14091/2/Cuaca-rata-rata-pada-Musim-gugur-di-Chicago-Illinois-Amerika-Serikat), **average summer temperatures** range from **20-28 degrees Celsius**, which is **warmer** than **other seasons** and Of course, the **lowest average temperature** occurs in **winter**, which is between **-6 to 1 degree Celsius**. That makes sense, as **warmer temperatures** are **suitable** for **cycling activities**, so the **busiest season** for **Cyclistic trips** is in the **summer**.
* Overall, **annual members** recorded **more total Cyclistic trip**s in **all seasons**.

### **Monthly**

````python
df_pv3 = pd.pivot_table(df_trips,
              index = ['ride_year', 'ride_month','ride_season'],
              values = 'ride_id',
              aggfunc = ['count'],
              margins = False)
df_pv3 = df_pv3.loc[(df_pv3 != 0).any(axis=1)]
df_pv3['MoM (%)']=df_pv3.pct_change()*100
df_pv3.style.background_gradient(cmap='YlGn', axis=0)
````

> **Insight**
* **The most Cyclistic user trips** occurred in **July 2022** or **mid-summer**, recording approximately **8.05 hundred trips** which corresponds to the **seasonal insight** that **summer** has **warmer temperatures** and is **suitable** for **cycling activities**.
* There is a **significant increase** in the number of **Cyclistic trips** by **141.74%** from **February (winter) to March (spring)** due to the **spring** season which has **warmer temperatures** than **winter**, so **Cyclistic activities increased more**.

````python
df_pv4 = pd.pivot_table(df_trips,
              index = ['ride_year', 'ride_month','ride_season'],
              columns='user_type',
              values = 'ride_id',
              aggfunc = ['count'],
              margins = False)
df_pv4 = df_pv4.loc[(df_pv4 != 0).all(axis=1)]
df_pv4.style.background_gradient(cmap='YlGn', axis=0)
````

> **Insight**
* **Annual members** make **more trips** than **casual riders** in **every month**.
